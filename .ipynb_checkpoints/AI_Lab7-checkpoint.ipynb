```python
def createMatrix(n):
    g=[]
    for i in range(n):
        r=[]
        for i in range(n):
            r.append(0)
        g.append(r)
    return g

def printGraph(G):
    for r in G:
        print(r)

f = open("g.txt","r")
vertex_edges=f.readline()
vertex_edges_list=vertex_edges.split()
number_of_vertex=int(vertex_edges_list[0])
number_of_edges=int(vertex_edges_list[1])
print(number_of_vertex,number_of_edges)
G=createMatrix(number_of_vertex)

for i in range(number_of_edges):
    edges=f.readline()
    edges_list=edges.split()
    u=int(edges_list[0])
    v=int(edges_list[1])
    G[u][v]=1

printGraph(G)

import queue
def dfs(G,s):
    stack=queue.LifoQueue()
    stack.put(s)
    explore_set=set()
    while not stack.empty():
        n=stack.get()
        print(n, end=" ")
        explore_set.add(n)
        for i in range(len(G[n])):
            if G[n][i]==1 and i not in explore_set:
                explore_set.add(i)
                stack.put(i)

dfs(G,0)
```

```python
import random
 
def createEmptyBoard(n):
    board = []
    for i in range(n):
        row = []  
        for j in range(n):
            row.append('.') 
        board.append(row)
    return board
 
def placeQueenOnBoard(B, row, col):
    B[row][col] = 'Q'
 
n = 6  
B = createEmptyBoard(n)  

for row in range(n):
    col = random.randint(0, n - 1) 
    placeQueenOnBoard(B, row, col) 
 

for row in B:
    print(' '.join(row)) 
 
```

```python
def createBoard(n):

    g=[]

    for i in range(n):

        r=[]

        for i in range(n):

           r.append(0)

        g.append(r)

    return g
 
def placeQueenOnBoard(B):

    for i in range(len(B)):

        randomNum = random.randint(0,len(B)-1)

        B[randomNum][i] = 1
 
def printBoard(G):

        for r in G:

         print(r)

 
 
def findString(B):

    string=""

    for i in range(len(B)):

        for j in range(len(B)):

            if B[j][i] == 1:

                string += str(j)

    print(string)

def createChromosome(n):

    string1=""

    for i in range(n):

        randomSt = random.randint(0,n-1)

        string1 += str(randomSt)

    return string1
 
def boardFromChromosome(chromosome):

    n = len(chromosome)

    board = createBoard(n)

    for col in range(n):

        row = int(chromosome[col])

        print(row)

        board[row][col] = 1

    return board
 
board = createBoard(8)

placeQueenOnBoard(board)

printBoard(board)

findString(board)

chromo = createChromosome(8)

print(chromo)
 
board1 = boardFromChromosome(chromo)

printBoard(board1)
 
 
```
